```python
prop = props[props.prop_name == 'PROPOSITION 062- DEATH PENALTY. INITIATIVE STATUTE.']
```

```python
prop.head()
```

```python
prop.info()
```

```python
contribs.info()
```

```python
prop.info()
```

```python
merged = pd.merge(prop, contribs, on="calaccess_committee_id")
```

ABOVE: Merging the prop and contribs DataFrames on the shared identifier "calaccess_committee_id" — defaults to an "inner" join, where you only get the values that match. That merge is saved to the DataFrame "merged."

```python
merged.head()
```

```python
merged.info()
```

BELOW: We will interview the merged DataFrame to find out who the biggest donors to Prop. 64 are. First question: What is the total amount of contributions to Prop. 64?

```python
merged.amount.sum()
```

This number, $35,177,017.64, doesn't account for donations smaller than $100, because those don't have to be documented in California. This total amount is how much the campaign received from "large donors."

Next question: How much did each side receive in contributions?

```python
merged.committee_position.value_counts()
```

```python
support = merged[merged.committee_position == 'SUPPORT']
```

```python
oppose = merged[merged.committee_position == 'OPPOSE']
```

```python
support.amount.sum()
```

```python
oppose.amount.sum()
```

```python
support.amount.sum() / merged.amount.sum()
```

By looking at the sum amounts of contributions for support and oppose, we can clearly see that support is larger. Using the divison operator and two sum operations, we can see that support makes up 92 percent of contributions for Prop. 64.

Third question: What were the ten biggest contributions to Prop. 64?

```python
merged.sort_values("amount", ascending=False)
```

```python
merged.sort_values("amount", ascending=False).head()
```

```python
support.sort_values("amount", ascending=False).head()
```

```python
oppose.sort_values("amount", ascending=False).head()
```

Above, we're able to see the top five biggest contributors to the support and opposition side. Chaining .head() to the end of the line of code lets us see the top five at a time.

```python
merged.groupby("committee_name_x").amount.sum()
```

```python
merged.groupby("committee_name_x").amount.sum().reset_index()
```

```python
merged.groupby("committee_name_x").amount.sum().reset_index().sort_values("amount", ascending=False)
```

ABOVE: You can group columns of data using groupby -- in this, we can see the top contributors to the proposition. (They're both support and oppose.)

```python
merged.groupby(["contributor_firstname", "contributor_lastname"]).amount.sum().reset_index().sort_values("amount", ascending=False).head()
```

```python
merged.groupby(["contributor_firstname", "contributor_lastname", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False).head()
```

Above, we looked at individual contributors to the campaign from most contributions to least.
